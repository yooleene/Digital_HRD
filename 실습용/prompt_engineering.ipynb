- in-Context Learning : 문맥(맥락)에 따라 학습

```python
'''
너는 리처드 파인만이고 나는 물리학에 대해 아무것도 모르는 일반인이다.
일반인도 이해하기 쉽게 불활정성 원리에 대해 200자 이내로 사례를 들어서 설명 해 줘라.
'''
```

물리학에서의 불확정성의 원리를 이해하기 쉽게 예시로 설명해 드리겠습니다. 생각해보세요, 우리가 동전을 던진다고 가정해봅시다. 동전이 떨어질 때, 우리는 '앞면' 또는 '뒷면' 중 하나가 나오기를 기대합니다. 하지만 불확정성의 원리에 따라 우리는 정확하게 어떤 면이 나올지 예측할 수 없습니다. 왜냐하면 우리는 던질 때 동전의 초기 상태, 공기 저항 등 많은 요인들을 정확하게 측정할 수 없기 때문입니다. 따라서 우리는 동전이 앞면과 뒷면 중 어느 면이 나올지, 어떤 확률로 결정될지를 예측합니다. 이와 같이 물리적 현상에서 불확정성의 원리는 미래의 결과를 정확하게 예측할 수 없음을 설명하고 있습니다.

```python
'''
여자친구가 평소에 내가 예쁘다고 생각하는 여자 연예인의 사진을 보여주면서 예쁘냐고 물어 봤다.
사실 그 연예인은 여자친구보다 더 예쁘다고 생각한다.
이때 나는 어떻게 대답해야 하는가? 
'''
```

이런 상황에서는 솔직하고 예의를 갖춘 대답을 하는 것이 좋습니다. 상대방의 민감한 감정을 생각하면서 다음과 같이 대답해 볼 수 있습니다:

"너도 아름답고 특별한 사람인데, 나는 그 사진에는 연예인의 매력을 볼 수 있지만 너에게서는 독특하고 사랑스러운 매력을 느껴요. 네가 나에게 가장 예쁘고 소중한 존재야." 

이렇게 해서 상대방의 자존감을 훼손하지 않으면서도 진심을 전할 수 있습니다. 상대방과의 솔직하고 강건한 관계를 유지하는 데 가장 중요한 것은 상대방을 존중하고 어필하는 것입니다.

```python
'''
다음 글을 참고하여 홍길동이 밥을 언제 먹었는지 알려줘라.

"""
새벽녘, 홍길동은 쌀로 만든 따뜻한 국물 요리와 신선한 야채 샐러드로 아침을 시작했다.
향긋한 녹차로 마무리 지으며 하루를 여는 에너지를 얻었다.
정오가 되자, 그는 해산물과 여러 채소가 들어간 김치찌개와 바삭한 김 몇 조각을 즐겼다.
오후 시간에는 과일과 견과류로 달콤하고 건강한 간식을 골랐다. 
저녁이 되어, 홍길동은 구운 닭고기와 다양한 야채를 곁들인 샐러드를 택했다. 
밤이 깇어감에 따라, 그는 가벼운 야식으로 우유와 쿠키를 조금 취했다. 
이렇게 홍길동은 하루 종일 다양한 음식을 취하며 강하고 활기찬 하루를 보냈다.
"""
''' 
```

주어진 글을 토대로 홍길동이 밥을 먹은 시간을 알아보면 다음과 같습니다:

- 아침: 새벽녘에 따뜻한 국물 요리와 야채 샐러드로 아침을 시작하였습니다.
- 정오: 해산물과 채소가 들어간 김치찌개와 김을 즐겼습니다.
- 오후: 과일과 견과류를 선택하여 간식을 즐겼습니다.
- 저녁: 구운 닭고기와 야채를 곁들인 샐러드를 선택했습니다.
- 밤: 우유와 쿠키를 야식으로 조금 취했습니다.

주어진 정보에 따라, 홍길동은 아침, 정오, 오후, 저녁, 밤의 다양한 시간에 음식을 취하였습니다.

- Step-by-Step Reasoning : 단계별 추론

```python
'''
다음 문제를 보고, 답을 말해줘. 단계별로 생각해 봅시다. 

"""
나는 시장에 가서 사과 10개게, 2개는 수리공를 샀습니다.
사과 2개는 이웃에에게 주었습니다.
그리고 가서 사과 5개를 더 사서 1개를 먹었습니다.
나에게는 몇 개의 사과가 남아 있었나요?
"""
'''
```

주어진 문제를 단계별로 생각해보겠습니다:

1. 처음에 시장에 가서 10개의 사과를 샀습니다. 
2. 그 중 2개를 이웃에게 주었으므로, 남은 사과는 10 - 2 = 8개입니다.
3. 또 다른 2개를 수리공에게 주었으므로, 남은 사과는 8 - 2 = 6개입니다.
4. 추가로 5개의 사과를 더 사서, 전체 사과의 개수는 6 + 5 = 11개입니다.
5. 그리고 1개의 사과를 먹었으므로, 남은 사과는 11 - 1 = 10개입니다.

따라서, 최종적으로 나에게는 10개의 사과가 남게 됩니다.

## 2. 프롬프트 엔지니어링(Prompt Engineering)

### 2-1. 프롬프트

#### 2-1-1. 프롬프트란? : 모델에게 입력하는 입력값

#### 2-1-2. 프롬프트의 요소
- Instruction : 모델에게 어떤 작업을 수행할 것인지 알려주는 것
- Context : 모델에게 어떤 상황에서 작업을 수행할 것인지 알려주는 것
- Input Data : 모델에게 어떤 데이터를 입력할 것인지 알려주는 것
- Output Indicator : 모델에게 어떤 결과를 출력할 것인지 알려주는 것

### 2-2. 프롬프트 엔지니어링

##### 2-2-1. 프롬프트 엔지니어링이란?

- 파인튜닝과 프롬프트 엔지니어링은 기계학습 기반의 자연어 처리 모델을 최적화하는 기술
- 이러한 기술은 이미 학습된 모델에 새로운 데이터를 반영하여 성능을 향상시키는 데 사용
- 파인튜닝은 기존 모델의 가중치를 조정하여 새로운 작업에 맞게 모델을 조정하는 것을 의미하며, 프롬프트 엔지니어링은 모델의 입력에 대한 적절한 프롬프트를 설계하여 모델의 출력을 개선하는 것을 의미

##### 2-2-2. 프롬프트 엔지니어링의 목적

- 파인튜닝은은 대규모의 텍스트 데이터세트를 사용하여 모델을 학습시킨 뒤, 특정 작업에 대해 추가로 학습시키기 때문에 높은 성능을 제공할 수 있으나 학습 시간과 비용, 많은 자원이 증가할 수 있음.
- 프롬프트 엔지니어링은 입력을 최적화하여 모델의 성능을 개선하기 때문에 많은 자원,시간,비용 등을 절감하면서 빠르게 모델의 성능을 개선할 수 있음.
- 환각(Hallucination) 현상을 방지할 수 있음.

하지만, 양자 컴퓨터는 아직까지는 실험실 수준에서 개발 중이고, 기술적인 어려움과 안정성의 문제를 해결해야 한다는 점을 알아두세요.

하지만 우리가 지금 사용하는 컴퓨터와 달리, 양자 컴퓨터는 아직 실험단계에 있고 개발이 많이 되어야 하니까 아직은 마법같은 컴퓨터에요.

```python
import random

def score():
    students = []  # 학생들의 정보를 저장할 리스트
    names = ["Alice", "Bob", "Charlie", "David", "Eve", "Frank", "Grace", "Harry", "Ivy", "Jack"]  # 학생들의 성명 리스트
    
    for i in range(10):
        name = random.choice(names)  # 임의로 학생의 성명 선택
        korean = random.randint(0, 100)  # 국어 성적 랜덤 생성
        english = random.randint(0, 100)  # 영어 성적 랜덤 생성
        math = random.randint(0, 100)  # 수학 성적 랜덤 생성
        
        student = {"name": name, "korean": korean, "english": english, "math": math}
        students.append(student)
    
    return students
```

### 3-1. Zero-shot Prompting

- 추가 학습 또는 예제 데이터 없이 답변을 생성하게 하는 것으로, 거대 언어 모델에게 아무런 지시 없이 완수할 태스크를 주는 방법

```python
'''
다음 문장의 긍정과 부정을 판단하세요.
본문: 오늘 날씨는 흐립니다.
'''
```

주어진 문장 "오늘 날씨는 흐립니다."는 일반적으로 날씨가 흐린 것을 나타내는 문장입니다. 일반적으로 날씨가 흐린 것은 긍정적인 의미를 담고 있지 않으며, 약간의 부정적인 뉘앙스를 가질 수 있습니다. 따라서, 이 문장은 부정적인 문장으로 판단될 수 있습니다.

```python
'''
다음 문장을 긍정과 부정으로 판단 할 수 있도록 파이썬으로 작성하세요.
본문: 오늘 날씨는 흐립니다. 
'''
```

```python
def sentiment_analysis(sentence):
    positive_words = ["좋다", "행복하다", "기쁘다"]  # 긍정적인 단어들의 리스트
    negative_words = ["나쁘다", "슬프다", "힘들다"]  # 부정적인 단어들의 리스트

    # 문장에서 긍정적인 단어가 있는지 확인
    for word in positive_words:
        if word in sentence:
            return "긍정"

    # 문장에서 부정적인 단어가 있는지 확인
    for word in negative_words:
        if word in sentence:
            return "부정"

    # 긍정적인 단어와 부정적인 단어가 모두 없을 경우, 중립으로 판단
    return "중립"
```

```python
sentence = "오늘 날씨는 흐립니다."
sentiment = sentiment_analysis(sentence)
print(sentiment)
```
